## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un",\
           "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"] 
# Diccionario para apendices
indices_duplicados = {}
# Bucle
for i, elemento in enumerate(ejer_3):
    elemento_minus = elemento.lower()
    if elemento_minus in indices_duplicados:
        indices_duplicados[elemento_minus].append(i)
    else:
        indices_duplicados[elemento_minus] = [i]
# Mostramos resultado
print("Los elementos duplicados y sus indices:")
for elemento, idxs in indices_duplicados.items():
    if len(idxs) > 1:
        print(f" '{elemento}') indices {idxs}")
`````

## Ejercicio 9
1. Escribe un programa que dada una variable titulo busque los libros que hay en la librería con ese título. Si no se encuentra debe devolver el mensaje `Ese no lo tengo, ¿mola?`
2. Pruena que el programa con uno de tus libros y con otro que no contenga tu librería.

`````python
libreria = [
    {"titulo": "Internet del dinero", "autor": "Antonopoulos"},
    {"titulo": "Padre rico padre pobre", "autor": "Robert Kiyosaki"},
    {"titulo": "El universo en una cáscara de nuez", "autor": "Stephen Hawkings"},
    {"titulo": "Aprendiz de brujo", "autor": "David Bronstein"}
]

def buscar_libro(titulo):
    titulo_buscar = titulo.lower()
    for libro in libreria:
        if titulo_buscar in libro["titulo"].lower():
            return libro
    return None

# Programa principal
titulo = input("Introduce el título del libro: ").strip()

if buscar_libro(titulo):
    libro = buscar_libro(titulo)
    print(f"¡Encontré el libro! '{libro['titulo']}' de {libro['autor']}")
else:
    print("Ese no lo tengo, ¿mola?")
`````

## Ejercicio 11
Convierte el programa del ejercicio 9 en una función al que se le pasen como argumentos la librería, posicional, y el título, argumento keyword con valor "ninguno", por defecto.

`````python
def buscar_libro_en_libreria(libreria, *, titulo="ninguno"):
    if titulo == "ninguno":
        return None
    titulo_buscar = titulo.lower()
    for libro in libreria:
        if titulo_buscar in libro["titulo"].lower():
            return libro
    return None

# Libreria
libreria = [
    {"titulo": "Internet del dinero", "autor": "Robert Kiyosaki"},
    {"titulo": "Aprendiz de Brujo", "autor": "David Bronstein"}
]

# Keyword
libro = buscar_libro_en_libreria(libreria, titulo="Mis geniales predecesores")
if libro:
    print(f"¡Encontrado! '{libro['titulo']}'")
else:
    print("Ese no lo tengo, ¿mola?")
`````

`````python
# Parte en la que se filtra la @
def es_email_valido(email):
    if "@" in email:
        return True
    else:
        return False
# Input para solicitar el mail al usuario

email_user = input("Por favor introduce tu dirección de email: ")

# Validación del correo y return

if es_email_valido(email_user):
    print((email_user),"El email es correcto!")
else:
    print((email_user),"¿Me estás intentando engañar? Ese mail no es correcto, falta @")
    
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
palabra = 'TRWAGMYFPDXBNJZSQVHLCE'

def es_dni_valido(dni):

    # Longitud
    print(f"Longitud del dni = {len(dni)}")
    if len(dni) <= 7 or len(dni) > 9:
        return False
    
    # 
    letra = dni[-1].upper()
    numeros = dni[:-1]
    print(f" La letra es ='{letra}', Y los numeros son ='{numeros}'")
    
    # 3. Longitud números
    print(f"La longitud de los numeros = {len(numeros)}")
    if len(numeros) <= 6 or len(numeros) > 8:
        return False
    
    # 4. REVISAR cada carácter (lo más básico)
    todos_numeros = True
    for caracter in numeros:
        if caracter not in "0123456789":
            print(f"'{caracter}' No es número!")
            todos_numeros = False
            break
    
    if todos_numeros == False:
        print(" No todos son números")
        return False
    
    # 5. Calcular letra correcta
    numero = int(numeros)
    letra_correcta = palabra[numero % 23]
    print(f" {numero} % 23 = {numero % 23}, '{letra_correcta}'")
    
    print(f"RESULTADO: '{letra}' No es igual a '{letra_correcta}', {letra == letra_correcta}")
    return letra == letra_correcta

# Prueba
es_dni_valido("45892753A")
es_dni_valido("45892753W")
`````
